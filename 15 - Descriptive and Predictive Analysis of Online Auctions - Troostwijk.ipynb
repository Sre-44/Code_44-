```python
df_company['bid_datetime'] = pd.to_datetime(df_company['bid_datetime'])

# Filter rows where 'bid_datetime' is before 00:00:00 of 1st September 2023
cutoff_date = pd.to_datetime('2023-09-01 00:00:00')
df_company = df_company[df_company['bid_datetime'] < cutoff_date]

# Calculate TOX, TOE, and NOB for each fake buyer and lot ID
bid_metrics = df_company.groupby(['fake_buyer_id', 'auction_id']).agg(
    TOX=('bid_datetime', 'max'),
    TOE=('bid_datetime', 'min'),
    NOB=('bid_datetime', 'count')
).reset_index()

# Merge these metrics back into the original DataFrame
df_company = df_company.merge(bid_metrics, on=['fake_buyer_id', 'auction_id'], how='left')

# Count unique auctions for each buyer
df_company['Number_of_Auctions'] = df_company.groupby('fake_buyer_id')['auction_id'].transform('nunique')

# For TLA, calculate only for those who have participated in more than one auction
# First, mark buyers with more than one auction
df_company['Participated_in_Multi_Auctions'] = df_company['fake_buyer_id'].isin(df_company[df_company['Number_of_Auctions'] > 1]['fake_buyer_id'])

# Sort by buyer and bid datetime to prepare for TLA calculation
df_company = df_company.sort_values(by=['fake_buyer_id', 'bid_datetime'])

# Calculate TLA only for those who participated in more than one auction
df_company.loc[df_company['Participated_in_Multi_Auctions'], 'prev_bid_datetime'] = df_company.groupby('fake_buyer_id')['bid_datetime'].shift(1)
df_company.loc[df_company['Participated_in_Multi_Auctions'], 'TLA'] = (df_company['bid_datetime'] - df_company['prev_bid_datetime']).dt.days

# Fill NaN in TLA for buyers who participated in only one auction with -1 (or any other placeholder)
df_company['TLA'] = df_company['TLA'].fillna(-1)

# Drop intermediate columns if not needed
df_company.drop(['prev_bid_datetime', 'Participated_in_Multi_Auctions'], axis=1, inplace=True)

# Display the first 30 rows
df_company.head(30)
```

```python
# Convert 'auction_start_time' to datetime without timezone
df_company['auction_start_time'] = pd.to_datetime(df_company['auction_start_time']).dt.tz_localize(None)
df_company['auction_close_time'] = pd.to_datetime(df_company['auction_close_time']).dt.tz_localize(None)
    
# Convert 'TOE' and 'TOX' to datetime without timezone
df_company['TOE'] = df_company['TOE'].dt.tz_localize(None)
df_company['TOX'] = df_company['TOX'].dt.tz_localize(None)

# Now perform the subtraction
df_company['TOE_seconds'] = (df_company['TOE'] - df_company['auction_start_time']).dt.total_seconds()
df_company['TOX_seconds'] = (df_company['TOX'] - df_company['auction_start_time']).dt.total_seconds()
df_company['TOA_seconds'] = (df_company['auction_close_time']- df_company['auction_start_time']).dt.total_seconds()
df_company['TOE_ratio'] = (df_company['TOE_seconds']/df_company['TOA_seconds'])
df_company['TOX_ratio'] = (df_company['TOX_seconds']/df_company['TOA_seconds'])
print(df_company.head(30))
```

# RFM-based Features

In the following cells, we do not create the traditional frequency, recency and monetary values, but their restriction to 3-month time windows. We plan to use RFM-related features from the past three months to predict Customer Lifetime Value (CLV) for the next month. For example, we will use features from months 1, 2, and 3 to train the model to predict the CLV for month 4. 
We have the CLV of months 4, 5, 6, 7 and 8 as the targets during the training and we have the CLV of month 9 as the target for the test. The reason why we are not including the CLV of months 1, 2, 3 as the targets during the training is we are using the data from 3 months prior to create the RFM-related features so it is not possible to create features for those months. You may wonder why we do not split the training and test sets to aviod data leakage. The reason is that we are creating data from training set to use in the test set.
During the feature creation, we avoid data leakage by not using the unknown data of month 9 in creating the features. we will only calculate the CLV of fake buyers in month 9 which is the target of the test set.

We will exclude the month 10 since we have data for only the first half of this month, meaning that we do not know the real added values of the customers over the entire month.

```python
# Filter the data to exclude the 10th month

df = df[df['month'] != 10]
df
```

We first calculate the recency of the customers at the end of each month separately, excluding the month 9 which is the testing period. Then, we use those columns to create three features called recency_last_month, rececny_last_second month, recency_last_third_month for each fake buyer in each month. At the end of feature creation, features we created for frecency of each month will be removed and only the final 3 features will be kept to implement the model.

```python
# calculate the recency at the end of 1st month

# Convert 'bid_datetime' to datetime 
df['bid_datetime'] = pd.to_datetime(df['bid_datetime'])

# Filter the dataframe to include only rows up to the end of month 1
df_up_to_month_1 = df[df['bid_datetime'] <= '2023-01-31']

# Find the most recent bid_datetime for each fake_buyer_id within the filtered data
max_bid_time_bidder_up_to_month_1 = df_up_to_month_1.groupby('fake_buyer_id')['bid_datetime'].transform('max')

# Assign this value to the original dataframe
df['max_bid_time_bidder_up_to_month_1'] = df['fake_buyer_id'].map(max_bid_time_bidder_up_to_month_1)

# Define the reference point as the end of month 1
end_of_month_1 = pd.Timestamp('2023-01-31')

# Calculate the difference and store in the new Recency column
df['Recency_end_of_month_1'] = end_of_month_1 - df['max_bid_time_bidder_up_to_month_1']

# Define the start of the year
start_of_year = pd.Timestamp('2023-01-01')

# Calculate the difference between the end of month 1 and the start of the year
recency_for_NaT = end_of_month_1 - start_of_year

# Fill NaT values in the Recency_end_of_month_1 column with this calculated difference
df['Recency_end_of_month_1'] = df['Recency_end_of_month_1'].fillna(recency_for_NaT)

# Drop unnecessary column
df.drop('max_bid_time_bidder_up_to_month_1', axis=1, inplace=True)
```

We now have the customer recency data for each month. The next step is to transform this data by creating three new columns: recency_last_month, recency_last_2nd_month, and recency_last_3rd_month. These columns will replace the existing recency data columns. Each row will then directly display the recency information for the last three months. The goal is that when training our model, we will analyze the relationship between the recency of the past three months and the Customer Lifetime Value (CLV) for the current month. It is important to note that the data leakage will not happen since we are using the recency values from the months 6, 7 and 8 to predict the CLV in month 9 (test period).

```python

# Initialize the new columns with NaT
df['recency_last_month'] = pd.NaT
df['recency_last_2nd_month'] = pd.NaT
df['recency_last_3rd_month'] = pd.NaT

# Define a function to get the recency columns
def get_recency_columns(row):
    month = row['month']
    if month in range(4, 10):
        row['recency_last_month'] = row.get(f'Recency_end_of_month_{month - 1}', pd.NaT)
        row['recency_last_2nd_month'] = row.get(f'Recency_end_of_month_{month - 2}', pd.NaT)
        row['recency_last_3rd_month'] = row.get(f'Recency_end_of_month_{month - 3}', pd.NaT)
    elif month in range(1, 4):
        # Since we are using 3-month time window to create features, we can not create features 
        # based on the last 3 months for months 1, 2 and 3. We are using the data of the first 
        # 3 months to create features for the upcoming months. for now we just assign a value of 0
        # and we will finally remove the rows associated with bids happening the first 3 months.
        
        row['recency_last_month'] = pd.Timedelta(days=0)
        row['recency_last_2nd_month'] = pd.Timedelta(days=0)
        row['recency_last_3rd_month'] = pd.Timedelta(days=0)
    return row

# Apply the function to each row
df = df.apply(get_recency_columns, axis=1)
df
```

Similar to topic 1, we first need to determine frequency to cacluate the CLV. This frequency is different from the frquency concept used in RFM. This is a part of the formula to calculate the CLV which is extracted from the paper. To compute frequency, we first need to compute the inter bidtime and then average interbidtime.

```python
# Compute Average inter bid time

import calendar

# Function to get the number of days in a month
def get_days_in_month(year, month):
    return calendar.monthrange(year, month)[1]

# Loop through each month and perform the operations
for month in range(1, 10):
    # Filter the dataframe to include only rows for the current month
    df_month = df[df['month'] == month]
    
    # Get the number of days in the current month
    days_in_month = get_days_in_month(2023, month)
    
    # Sort the DataFrame by fake_buyer_id and bid_datetime
    df_month = df_month.sort_values(by=['fake_buyer_id', 'bid_datetime'])
    
    # Calculate the interbidtime in hours (If x is NaT (which means there is no previous bid), we assign the total number of hours in the month, effectively assuming the interbid time is the entire month if it is the only bid of the fake_buyer_id in that month.)
    df_month['interbidtime'] = df_month.groupby('fake_buyer_id')['bid_datetime'].diff().apply(lambda x: x.total_seconds() / 3600 if pd.notnull(x) else days_in_month * 24)
    
    # Calculate the average interbidtime for each fake_buyer_id
    average_interbidtime = df_month.groupby('fake_buyer_id')['interbidtime'].mean().reset_index()
    average_interbidtime.columns = ['fake_buyer_id', f'mean_interbidtime_month_{month}']
    
    # Merge the average interbidtime back into the original DataFrame
    df = df.merge(average_interbidtime, on='fake_buyer_id', how='left')
    
```

There is a difference between calculating the monthly vs yearly CLV. We used the yearly discount rate to calculate yearly CLV but here we should devide that by 12 to have a monthly discount rate. Plus, instead of considering the time value of the months, we consider the time value of the days.

```python
# Calculate the first part of the formula without sum for the given month

# Convert 'bid_datetime' to datetime
df['bid_datetime'] = pd.to_datetime(df['bid_datetime'])

# Extract day and month from 'bid_datetime'
df['day'] = df['bid_datetime'].dt.day
df['month'] = df['bid_datetime'].dt.month


# Calculate 'First_Part_of_the_Formula_without_sum' for each row
df['First_Part_of_the_Formula_without_sum'] = (
    df['added_value_weighted'] * df['freq']
) / np.power((1 + 0.022/12), (31 - df['day'] / 31))

```

```python
# Calculate the first part of the formula 

# Initialize a dictionary to hold the DataFrames for each month
formula_dict = {}

# Loop through each month and calculate the first part of the formula values
for month in range(1, 10):
    formula_col = f'First_Part_of_the_Formula_month_{month}'
    # Group by 'fake_buyer_id' and 'month', and sum 'First_Part_of_the_Formula_without_sum'
    formula_monthly = df[df['month'] == month].groupby('fake_buyer_id')['First_Part_of_the_Formula_without_sum'].sum().reset_index()
    formula_monthly.columns = ['fake_buyer_id', formula_col]
    
    # Store the monthly DataFrame in the dictionary
    formula_dict[month] = formula_monthly

# Merge all the monthly DataFrames into the original DataFrame
for month, formula_df in formula_dict.items():
    df = pd.merge(df, formula_df, on='fake_buyer_id', how='left')

# Fill NaN values in the new columns with 0
for month in range(1, 10):
    formula_col = f'First_Part_of_the_Formula_month_{month}'
    df[formula_col] = df[formula_col].fillna(0)

```

```python
# Convert bid_datetime to datetime
df_c['bid_datetime'] = pd.to_datetime(df_c['bid_datetime'])

# Filter to get the minimum and maximum bid datetime within the 7th month (July)
min_bid_datetime = df_c['bid_datetime'].min()
max_bid_datetime_7th_month = df_c['bid_datetime'].max()

# Create a new column 'weeks' based on the difference between the bid_datetime and the minimum bid datetime
df_c['weeks'] = ((df_c['bid_datetime'] - min_bid_datetime).dt.days // 7) + 1

# Filter weeks_c_complete to include only the weeks within the desired timeframe
weeks_c_complete = df_c[(df_c['bid_datetime'] >= min_bid_datetime) & (df_c['bid_datetime'] <= max_bid_datetime_7th_month)]

# Group by fake_buyer_id and weeks, then sum the added_value and take the mean of interbidtime
weeks_c_added_value = weeks_c_complete.groupby(['fake_buyer_id', 'weeks'])['added_value_weighted'].sum().reset_index()
weeks_c_interbidtime = weeks_c_complete.groupby(['fake_buyer_id', 'weeks'])['interbidtime'].mean().reset_index()

# Merge the two dataframes on fake_buyer_id and weeks
weeks_c = pd.merge(weeks_c_added_value, weeks_c_interbidtime, on=['fake_buyer_id', 'weeks'])

# Generate a complete list of weeks for each fake_buyer_id
all_weeks = pd.MultiIndex.from_product(
    [weeks_c_complete['fake_buyer_id'].unique(), range(1, weeks_c['weeks'].max() + 1)],
    names=['fake_buyer_id', 'weeks']
).to_frame(index=False)

# Merge all weeks with the grouped data to ensure all weeks are included
weeks_c_complete = pd.merge(all_weeks, weeks_c, on=['fake_buyer_id', 'weeks'], how='left')

# Fill NaN values with 0
weeks_c_complete.fillna(0, inplace=True)
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# Calculate the frequency column as 1/interbidtime
weeks_c_complete['frequency'] = np.where(weeks_c_complete['interbidtime'] != 0,
                                         1 / weeks_c_complete['interbidtime'],
                                         np.inf)

# Replace infinity values with the minimum non-infinite frequency
min_frequency = weeks_c_complete.replace([np.inf, -np.inf], np.nan)['frequency'].min()
weeks_c_complete['frequency'].replace(np.inf, min_frequency, inplace=True)

# Scale the frequency column between 1 and 10
scaler = MinMaxScaler(feature_range=(1, 10))
weeks_c_complete['frequency'] = scaler.fit_transform(weeks_c_complete[['frequency']])
# Create data1 as weeks_c_complete
data1 = weeks_c_complete

# Create data2 as a DataFrame with fake_buyer_id and marketing_cost_per_customer from df_c
data2 = df_c[['fake_buyer_id', 'marketing cost per customer']].drop_duplicates()

# Merge data1 and data2 on 'fake_buyer_id' using a left join
output2 = pd.merge(data1, data2, on='fake_buyer_id', how='left')

# Update weeks_c_complete with the merged result
weeks_c_complete = output2
# Calculate the discount column using the formula np.power(1.022, (52 - weeks_c_complete['weeks'] / 52))
weeks_c_complete['discount'] = np.power(1.022, (52 - weeks_c_complete['weeks']) / 52)
# Calculate partial_CLV column
weeks_c_complete['partial_CLV'] = (weeks_c_complete['added_value_weighted'] * weeks_c_complete['frequency']) / weeks_c_complete['discount']
# Calculate full_CLV_1 column using cumulative sum grouped by fake_buyer_id
weeks_c_complete['full_CLV_1'] = weeks_c_complete.groupby('fake_buyer_id')['partial_CLV'].cumsum()
weeks_c_complete['CLV']  = weeks_c_complete['full_CLV_1'] - weeks_c_complete['marketing cost per customer']
```
